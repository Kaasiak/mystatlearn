# Kerndel density estimation

```python
from scipy.stats import norm

n = 100
sample = np.hstack([norm.rvs(size=n), norm.rvs(size=n, loc=3)])

kde = KDE(kernel='gaussian', bandwidth=0.5)
kde.fit(sample)

x = np.linspace(-5, 8, 100)

fig, ax = plt.subplots()
ax.plot(x, norm.pdf(x) * 0.5 + norm.pdf(x, loc=3) * 0.5, linestyle='--', alpha=0.5, label='true density $f$', color='C4')
ax.plot(x, kde.density(x), color='C4', label = 'kde $\hat{f}_n$')
ax.plot(sample, [0.01]*len(sample), '|', color='k')
ax.legend()
plt.show()
```
